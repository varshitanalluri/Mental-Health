```python
import pandas as pd
import random
import faker

fake = faker.Faker()

samples_per_level = 3000  
total_samples = samples_per_level * 3 

response_mapping = {
    "Not at all": 0,
    "Several days": 1,
    "More than half the days": 2,
    "Nearly every day": 3
}
responses = list(response_mapping.keys())

anxiety_questions = [
    "Feeling nervous, anxious or on edge",
    "Not being able to stop or control worrying",
    "Worrying too much about different things",
    "Trouble relaxing",
    "Being so restless that it is hard to sit still",
    "Becoming easily annoyed or irritable",
    "Feeling afraid as if something awful might happen"
]

stress_questions = [
    "Feeling tired or having little energy",
    "Poor appetite or overeating",
    "Feeling bad about yourselfâ€”or that you are a failure",
    "Trouble concentrating on things",
    "Moving or speaking so slowly that other people could have noticed",
    "Thoughts that you would be better off dead or of hurting yourself in some way"
]

def generate_responses(target_score):
    while True:
        responses_dict = {q: random.choice(responses) for q in anxiety_questions + stress_questions}
        total_score = sum(response_mapping[responses_dict[q]] for q in anxiety_questions)
        
        if total_score == target_score: 
            return responses_dict

def classify_level(score):
    if score <= 5:
        return "Low"
    elif 6 <= score <= 10:
        return "Moderate"
    else:
        return "High"

data = []
for level, score_range in zip(["Low", "Moderate", "High"], [(0, 5), (6, 10), (11, 21)]):
    for _ in range(samples_per_level):
        name = fake.name()
        email = fake.email()
        age = random.randint(17, 21)  


        target_anxiety_score = random.randint(*score_range)
        survey_responses = generate_responses(target_anxiety_score)
        
        target_stress_score = random.randint(*score_range)  

        data.append({
            "Name": name,
            "Email": email,
            "Age": age,
            **survey_responses,
            "Anxiety_Score": target_anxiety_score,
            "Stress_Score": target_stress_score,
            "Anxiety_Level": level,
            "Stress_Level": level
        })

df = pd.DataFrame(data)

output_file = "Children_Dataset.xlsx"
df.to_excel(output_file, index=False)

print(f"Dataset saved as {output_file}")
```
